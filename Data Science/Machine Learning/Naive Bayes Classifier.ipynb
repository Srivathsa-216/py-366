```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
target = df.Survived
inputs = df.drop('Survived', axis = 'columns')
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
dummies = dummies.rename(columns={'male': 'Gender'})
```

```python
dummies
```

```python
dummies.drop(columns=['female'], inplace=True)
```

```python
inputs = pd.concat([inputs, dummies] , axis = 'columns')
inputs.head()
```

```python
inputs.drop(columns=['Sex'], inplace=True)
```

```python
inputs.head()
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
len(inputs)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()    ### Gaussian NB is used when the data is normally distributed
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_test[:10]
```

```python
model.predict(x_test[:10])
```

```python
model.predict_proba(x_test[:10])
```

# Email spam detection

```python
import pandas as pd
df = pd.read_csv("Spam.csv")
df.head()
```

```python
df.groupby('Category').describe()
```

```python
encoded_df = pd.get_dummies(df['Category'], prefix='Category', drop_first=True)

# Concatenate the encoded columns with the original DataFrame
df = pd.concat([df, encoded_df], axis=1)
```

```python
df.head()
```

```python
df.drop(columns=['Category'], inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.Message, df.Category_spam, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
x_train_count = v.fit_transform(x_train.values)
x_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
x_test_count = v.transform(x_test)
model.score(x_test_count, y_test)
```

sklearn Pipeline

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```

```python
clf.predict(emails)
```
