### Home Price Prediction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("HousePriceData/HousePriceList.csv")
df.head()
```

```python
%matplotlib inline
```

```python
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.xlabel('area(sq ft)')
plt.ylabel('price(US $)')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3000]])   #### x is 3000
```

```python
df.head()
```

```python
reg.coef_   #### m
```

```python
reg.intercept_    #### b
```

```python
### y = mx +b

26.55821918 * 3000 + (-13410.95890410959)
```

```python
d = pd.read_csv('HousePriceData/predict.csv')
d.head()
```

```python
p = reg.predict(d)
```

```python
p
```

```python
d['price'] = p
```

```python
d.head()
```

```python
d.to_csv('HousePriceData/predicted_prices.csv',index = False)
```

```python
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.xlabel('area(sq ft)')
plt.ylabel('price(US $)')
plt.plot(df.area, reg.predict(df[['area']]), color = 'blue')
```

Finding per capita income of canada at the year 2020

```python
df = pd.read_csv('HousePriceData/Canada_perCapitaIncome.csv')
df.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[1970]])
```

```python
plt.scatter(df.year, df['per capita income (US$)'], color = 'red', marker = '+')
plt.xlabel('year')
plt.ylabel('Income(US $)')
plt.plot(df.year, reg.predict(df[['year']]), color = 'blue')
```
